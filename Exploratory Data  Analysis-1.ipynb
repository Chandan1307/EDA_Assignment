# Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.

```python
# Missing Value Check 
dataset.isnull().sum()
```

There are several techniques available for imputing missing values in a data set, each with its own advantages and disadvantages.

Mean/Median/Mode imputation:
This involves replacing missing values with the mean, median or mode of the non-missing values for that variable. This is a simple and fast method that can work well if the data is missing completely at random and the amount of missing data is small. However, it can lead to biased estimates if there is a systematic reason for the missing data and can also reduce the variance of the data.

Regression imputation:
This involves predicting the missing values from other variables in the data set using a regression model. This method can produce accurate imputations if the variables used in the model are highly predictive of the missing values. However, it can be computationally expensive and may not work well if there are strong nonlinear relationships between the variables.

K-nearest neighbor (KNN) imputation:
This involves finding the K closest observations to the missing value and using their values to impute the missing value. This method can be effective if there is a lot of missing data and the observations are highly correlated with each other. However, it can be computationally expensive and may not work well if there are many variables with missing data.

Multiple imputation:
This involves creating multiple imputed data sets using a model-based approach and then averaging the results. This method can produce accurate results if the imputation model is well-specified and there is a low amount of missing data. However, it can be computationally expensive and may not work well if the imputation model is poorly specified or the amount of missing data is high.

In general, the choice of imputation technique depends on the nature of the missing data and the goals of the analysis. It is important to carefully consider the advantages and disadvantages of each technique before making a decision.

# Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the wine quality dataset
wine_data = pd.read_csv('winequality-red.csv')

# Separate the features and target variable
X = wine_data.drop('quality', axis=1)
y = wine_data['quality']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Calculate explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Find the minimum number of principal components to explain 90% variance
n_components = np.argmax(cumulative_explained_variance >= 0.9) + 1

print(f"Minimum number of principal components to explain 90% variance: {n_components}")
```
